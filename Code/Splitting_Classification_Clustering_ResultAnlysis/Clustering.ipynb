# Load Dataset

```python
import pandas as pd
#Load Dataset
df = pd.read_csv('D:/WorkShop.csv')
df
```

```python
#Assign independent features into x
x = df.drop(['ID','Gender'],axis=1) # axis=0 will act on all the ROWS in each column. axis 1 will act on all the COLUMNS in each row.
x
```

```python
x
```

# Simple K means

```python
#Step_1: Import the KMeans
from sklearn.cluster import KMeans
#Step_2: Create cluster class
km = KMeans(n_clusters=3)
#Step_3: Fit and predict cluster
y_predicted_km = km.fit_predict(x)
df['cluster_SkM']=y_predicted_km
df
```

# Agglomerative Clustering

```python
#Step_1: Import the Agglomerative
from sklearn.cluster import AgglomerativeClustering
#Step_2: Create AgglomerativeCluster class
Agglo = AgglomerativeClustering()
#Step_3: Fit and predict cluster
y_predicted_agglo = Agglo.fit_predict(x)
df['Cluster_Aglo']=y_predicted_agglo 
df
```
