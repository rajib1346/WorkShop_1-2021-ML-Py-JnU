# Voting Ensemble Method with Cross Validation

```python
#Step_1: Import Cross Validation Library
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_predict

#Import metrics to calculate the score
from sklearn import metrics


#Step_2: Import Voting Classifier
from sklearn.ensemble import VotingClassifier

#Step_3: Import all base classifiers
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm

#Step_4: Create all base classifier
b_rf  = RandomForestClassifier()
b_knn = KNeighborsClassifier(n_neighbors=2)
b_nb  = GaussianNB()
b_dt = DecisionTreeClassifier()
b_sv = svm.SVC(probability=True)  #Enable probability.By default its disable

# AdaBoost
#XgBoost
#Gradient Boosting
#Logistic Regression

#[N.B: Source code is hidden. If you want to declare each of the above four as a base classifier please follow above steps 3 and 4.]


# Create VotingClassifier 
VC = VotingClassifier (estimators=[('cl1', b_rf), ('cl5', b_knn), ('cl7', b_nb), ('cl8', b_dt), 
                                    ('cl9', b_sv)], weights=[2,4,2,4,2], voting='soft')    
                                                            #You can also use hard voting
    
# Create a classifier list
def get_models():
#Create a dictionary containing all classfiers
    models = dict()
    models['Random Forest'] = b_rf 
    models['K Nearest Neighbor'] = b_knn
    models['Navie Bayes'] = b_nb
    models['Decision Tree'] = b_dt
    models['Support Vector Machine'] = b_sv
    models['Voting Classifer'] = VC
    return models   

# Get all classifiers from the classifier list
models = get_models()

predicts=list()
names=list()

# Create a classification block
def evaluate_model(model, X, y):
    #Create Stratified K Fold
    Sfolds = StratifiedKFold(n_splits=4, shuffle=True)
    predict = cross_val_predict(model, X, y, cv=Sfolds)
    return predict

print('-----------------Accuracy List-----------------')
# Pass all model and data to classification block
for name, model in models.items():
    predict = evaluate_model(model, x, y)
    predicts.append(predict)
    names.append(name)
    print('%s: ' %(name))
    print('Predicted Class: ', predict)
    print('Accuracy: %0.2f: ' %(metrics.accuracy_score(y, predict)*100))
```

```python
#Transpose the Row
trans = dataF.transpose()
trans
```

```python
#Rename the column name
NewFrame = trans.rename(columns={0:'RF', 1:'KNN', 2:'NB', 3:'DT', 4:'SVM', 5:'Voting'})
NewFrame
NewFrame['ActualClass']=y
NewFrame
```

# Confusion Matrix

```python
from sklearn import metrics

ConMat = metrics.confusion_matrix(NewFrame['ActualClass'], NewFrame['Voting'])
# Extract TN, FP, FN, TP
TN,FP,FN,TP = ConMat.ravel()
#Calculate Accuracy
Accuracy = metrics.accuracy_score(NewFrame['ActualClass'], NewFrame['Voting'])*100

# All Formula 
Prec=TP/(TP+FN)
TPR=TP/(TP+FN)
FPR=FP/(TN+FP)
TNR=TN/(TN+FP)
FNR=FN/(FN+TP)

#Print the result
print("Confusion Matrix:\n",ConMat)
print('True Negative:', TN)
print('False Positive:', FP)
print('False Negative:', FN)
print('True Positive:', TP)
print('Accuracy: %.2f' %Accuracy)
print('Precision: %.2f' %Prec)
print('Recall/Sensitivity/True Positive Rate: %.2f' %TPR)
print('False Positive Rate: %.2f' %FPR)
print('Specificity/True Negative Rate: %.2f' %TNR)
print('False Negative Rate: %.2f' %FNR)
```

# ROC and AUC

```python
from sklearn.metrics import roc_curve,auc
from matplotlib import pyplot as plt


fpr, tpr, treshold= roc_curve(NewFrame['ActualClass'], NewFrame['Voting'])
auc_sc=auc(fpr,tpr)

plt.figure(figsize=(5,5),dpi=100) #dpi = dots per inch
plt.plot(fpr,tpr,linestyle='-',label='model_auc (auc=%0.2f)' %auc_sc)
plt.plot([0,1],[0,1], color='orange',linestyle='--')

plt.xlabel('False Positive Rate-->')
plt.ylabel('True Positive Rate-->')
plt.legend()
plt.show()
```
