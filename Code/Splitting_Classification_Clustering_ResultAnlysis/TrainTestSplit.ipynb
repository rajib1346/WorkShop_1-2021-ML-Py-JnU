```python
#Load the train and test dataset
Train = pd.read_csv('D:/Train_D6.csv')
Test = pd.read_csv('D:/Test_D6.csv')
Train.shape, Test.shape
```

```python
#Assign the independent and dependent features of train dataset into X_train and Y_train respectively
X_train=Train.drop(['class'],axis=1) 
Y_train=Train['class']

#Assign the independent and dependent features of test dataset into X_test and Y_test respectively
X_test=Test.drop(['class'],axis=1) 
Y_test=Test['class']
```

# Ensemble Classification

```python
#Step_1: Import VotingClassifier
from sklearn.ensemble import VotingClassifier

#Step_2: Import all base classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm

#Step_3: Create all base classifier
b_rf  = RandomForestClassifier()
b_knn = KNeighborsClassifier(n_neighbors=2)
b_nb  = GaussianNB()
b_dt = DecisionTreeClassifier()
b_sv = svm.SVC(probability=True) #Enable probability.By default its disable

# AdaBoost
#XgBoost
#Gradient Boosting
#Logistic Regression

#[N.B: Source code is hidden. If you want to declare each of the above four as a base classifier please follow above steps 2 and 3.]

#Step_4: Create the VotingClassifier 
VC = VotingClassifier (estimators=[('cl1', b_rf), ('cl5', b_knn), ('cl7', b_nb), ('cl8', b_dt), 
                                   ('cl9', b_sv)], weights=[2,4,2,4,2], voting='hard') 
                                                                        # you can use hard voting or soft voting




#Step_5: Fit the Voting Classifer
VC.fit(X_train, Y_train)
#Step_6: Predict the target Class
y_predict_ensem = VC.predict(X_test)
#Step_5: Calculate the score
from sklearn import metrics
vc_acc = metrics.accuracy_score(Y_test, y_predict_ensem)*100

print("Voting Ensemble Classifer Accuracy: ",'%.2f' %vc_acc)
```

# Confusion Matrix

```python
from sklearn import metrics

# Build Confusion Matrix 
ConMat = metrics.confusion_matrix(Y_test, y_predict_ensem)

# Extract TN, FP, FN, TP from multidimentional array
TN,FP,FN,TP = metrics.confusion_matrix(Y_test,y_predict_ensem).ravel()

#Calculate Accuracy
Accuracy = metrics.accuracy_score(Y_test, y_predict_ensem)*100

# All Formula 
Prec=TP/(TP+FP)
TPR=TP/(TP+FN)
FPR=FP/(TN+FP)
TNR=TN/(TN+FP)
FNR=FN/(FN+TP)

#Print the result
print("Confusion Matrix:\n",ConMat)
print('True Negative:', TN)
print('False Positive:', FP)
print('False Negative:', FN)
print('True Positive:', TP)
print('Accuracy: %.2f' %Accuracy)
print('Precision: %.2f' %Prec)
print('Recall/Sensitivity/True Positive Rate: %.2f' %TPR)
print('False Positive Rate: %.2f' %FPR)
print('Specificity/True Negative Rate: %.2f' %TNR)
print('False Negative Rate: %.2f' %FNR)
```

# ROC and AUC

```python
from sklearn.metrics import roc_curve,auc
from matplotlib import pyplot as plt

fpr, tpr, treshold= roc_curve(Y_test,y_predict_ensem)

auc_sc=auc(fpr,tpr)

plt.figure(figsize=(5,5),dpi=100) #dpi = dots per inch
plt.plot(fpr,tpr,linestyle='-',label='model_auc (auc=%0.2f)' %auc_sc)
plt.plot([0,1],[0,1], color='orange',linestyle='--')

plt.xlabel('False Positive Rate-->')
plt.ylabel('True Positive Rate-->')
plt.legend()
plt.show()
```
