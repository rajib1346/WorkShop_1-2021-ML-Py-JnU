# Frequency Encoding

```python
import pandas as pd
data = pd.read_csv("H:/Machine Learning/Practical/Data_null_removed.csv")
data = data.drop(['Unnamed: 0'], axis=1)
data
```

```python
data.dtypes
```

```python
enc_rbc = (data.groupby(data['rbc']).size()) / len(data)
enc_rbc
```

```python
data['rbc_encode'] = data['rbc'].apply(lambda x : enc_rbc[x])
```

```python
data
```

```python
enc_pc = (data.groupby(data['pc']).size()) / len(data)
data['pc_encode'] = data['pc'].apply(lambda x : enc_pc[x])
enc_pcc = (data.groupby(data['pcc']).size()) / len(data)
data['pcc _encode'] = data['pcc'].apply(lambda x : enc_pcc[x])
enc_ba = (data.groupby(data['ba']).size()) / len(data)
data['ba_encode'] = data['ba'].apply(lambda x : enc_ba[x])
enc_htn = (data.groupby(data['htn']).size()) / len(data)
data['htn _encode'] = data['htn'].apply(lambda x : enc_htn[x])
enc_dm = (data.groupby(data['dm']).size()) / len(data)
data['dm_encode'] = data['dm'].apply(lambda x : enc_dm[x])
enc_cad = (data.groupby(data['cad']).size()) / len(data)
data['cad _encode'] = data['cad'].apply(lambda x : enc_cad[x])
enc_appet = (data.groupby(data['appet']).size()) / len(data)
data['appet_encode'] = data['appet'].apply(lambda x : enc_appet[x])
enc_pe = (data.groupby(data['pe']).size()) / len(data)
data['pe _encode'] = data['pe'].apply(lambda x : enc_pe[x])
enc_ane = (data.groupby(data['ane']).size()) / len(data)
data['ane_encode'] = data['ane'].apply(lambda x : enc_ane[x])
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr    
```

```python
corr_features = correlation(df_1, 0.8)
```
